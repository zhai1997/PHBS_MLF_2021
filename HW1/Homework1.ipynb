#### Import packages and data

```python
from scipy.io import arff
import pandas as pd
import numpy as np

data = arff.loadarff('4year.arff')
df = pd.DataFrame(data[0])
```

```python
df
```

#### Create a new dataframe

```python
newdf=df[['Attr1','Attr2','Attr7','Attr10','class']]
newdf
```

#### Rename the columns

```python
newdf.columns=['X1','X2','X7','X10','class']
newdf
```

#### Fill in missing values

```python
newdf=newdf.fillna(newdf.mean())
newdf
```

#### Mean and standard deviation

```python
newdf.groupby('class').agg(['mean','std'])
```

```python
newdf.drop('class',axis=1).agg(['mean','std'])
```

#### Subgroup that satisfy the condition

```python
subdf=newdf[(newdf.X1<newdf.X1.mean()-newdf.X1.std())&(newdf.X10<newdf.X10.mean()-newdf.X10.std())]
subdf
```
